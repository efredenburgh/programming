```python
# step 1
import numpy as np
import matplotlib.pyplot as plt

# step 2
grid = np.linspace(0, 1, 50)

# step 3
y = np.log(grid)
z = np.exp(grid)

# step 4/7
plt.scatter(grid, y, label='Natural Log')
plt.scatter(grid, z, label='Exponential')

# step 6
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Natural Log and Exponential Functions")

# step 7
plt.legend(loc = 'lower right')

# step 5
plt.show()



# doing it again "with slightly less direction"

new_grid = np.linspace(-6.5, 6.5, 100)

sin = np.sin(new_grid)
cos = np.cos(new_grid)

plt.plot(new_grid, sin, label='Sine')
plt.plot(new_grid, cos, label='Cosine')

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Sine & Cosine Functions")

plt.legend(loc = 'lower left')

plt.show()
```
